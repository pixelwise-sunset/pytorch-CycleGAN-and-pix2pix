```python
from PIL import Image

# 打开图片
img = Image.open("datasets\horse2zebra/testA/n02381460_20.jpg")

# 获取分辨率 (宽度, 高度)
width, height = img.size
print(f"分辨率: {width} x {height}")

# 获取 DPI（如果有）
dpi = img.info.get("dpi", (None, None))
print(f"DPI: {dpi}")
```

```python
import matplotlib.pyplot as plt

# 读取两张图片
img_fake = plt.imread('./results/low2high/test_latest/images/547_fake.png')
img_real = plt.imread('./results/low2high/test_latest/images/547_real.png')
# img_fake = plt.imread('./results/horse2zebra_pretrained/test_latest/images/n02381460_20_fake.png')
# img_real = plt.imread('./results/horse2zebra_pretrained/test_latest/images/n02381460_20_real.png')

# 创建一个1行2列的子图布局
fig, axes = plt.subplots(1, 2, figsize=(10, 5))  # figsize调整显示大小

# 分别显示 fake 和 real 图片
axes[0].imshow(img_fake)
axes[0].set_title('Fake')  # 添加标题
axes[0].axis('off')  # 关闭坐标轴

axes[1].imshow(img_real)
axes[1].set_title('Real')
axes[1].axis('off')

# 自动调整子图间距
plt.tight_layout()

# 显示图像
plt.show()
```
